Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.


Answer = Probability Mass Function (PMF) and Probability Density Function (PDF) are two fundamental concepts in probability theory that describe the likelihood of different outcomes for random variables.

Probability Mass Function (PMF)

A probability mass function (PMF) is a function that assigns probabilities to a discrete random variable. In other words, a PMF tells us the probability of each possible value of a discrete random variable.

Example:

Consider the random variable representing the number of heads obtained when flipping a coin twice. The possible values of this random variable are 0, 1, or 2. The PMF for this random variable is given by:

P(X = 0) = 1/4
P(X = 1) = 1/2
P(X = 2) = 1/4
This PMF tells us that the probability of getting 0 heads is 1/4, the probability of getting 1 head is 1/2, and the probability of getting 2 heads is 1/4.

Probability Density Function (PDF)

A probability density function (PDF) is a function that describes the probability distribution of a continuous random variable. Unlike the PMF, which assigns probabilities to discrete values, a PDF assigns probabilities to intervals of values.

Example:

Consider the random variable representing the height of a randomly selected adult. The possible values of this random variable are all real numbers between 4 feet and 7 feet. The PDF for this random variable cannot be represented as a simple list of probabilities because there are an infinite number of possible values. Instead, the PDF is represented by a function that gives the probability for any given interval of values.

For example, the probability of the height being between 5.5 feet and 6 feet can be calculated by integrating the PDF over that interval.

Key Differences

The main difference between the PMF and PDF is that the PMF applies to discrete random variables, while the PDF applies to continuous random variables. Additionally, the PMF assigns probabilities to individual values, while the PDF assigns probabilities to intervals of values.



Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

Answer = 
The cumulative distribution function (CDF) is a function that describes the probability that a random variable will take on a value less than or equal to a certain value. In other words, the CDF tells us the proportion of the total probability that lies below a certain threshold.

Definition

The CDF of a random variable X is defined as the function:

F(x) = P(X ≤ x)
where:

F(x) is the CDF of X
P(X ≤ x) is the probability that X will take on a value less than or equal to x
Example:

Consider the random variable representing the number of heads obtained when flipping a fair coin three times. The possible values of this random variable are 0, 1, 2, or 3. The CDF for this random variable is given by:

F(x) = 0 if x < 0
F(x) = 1/8 if 0 ≤ x < 1
F(x) = 3/8 if 1 ≤ x < 2
F(x) = 5/8 if 2 ≤ x < 3
F(x) = 1 if x ≥ 3
This CDF tells us that the probability of getting 0 heads is 0, the probability of getting 1 head is 1/8, the probability of getting 2 heads is 3/8, and so on.

Uses of CDF

The CDF is a versatile tool with various applications in probability theory and statistics. Here are some common uses of the CDF:

Calculating Probabilities: The CDF can be used to calculate the probability that a random variable will take on a value within a certain range. For example, the probability of getting between 1 and 2 heads when flipping a fair coin three times can be calculated as:
P(1 ≤ X ≤ 2) = F(2) - F(1) = 3/8 - 1/8 = 1/4
Comparing Distributions: The CDFs of different distributions can be compared to assess their relative dispersions. A steeper CDF indicates a more concentrated distribution, while a flatter CDF indicates a more spread-out distribution.

Generating Random Samples: The CDF can be used to generate random samples from a distribution. This is often done using the inverse transform sampling method.

Hypothesis Testing: The CDF is used in various statistical tests to determine the p-value, which is a measure of the strength of evidence against a null hypothesis.

Risk Assessment: The CDF can be used to assess the risk associated with a particular event. For example, it can be used to calculate the probability of a stock price falling below a certain threshold.

Answer = Bernoulli Distribution

The Bernoulli distribution is characterized by a single parameter, p, which represents the probability of success on a single binary trial.

The probability of success is denoted by p, and the probability of failure is denoted by 1 - p.

The Bernoulli distribution is used to model situations where there are only two possible outcomes, such as flipping a coin (heads or tails) or rolling a die (even or odd).

The expected value (mean) of the Bernoulli distribution is given by:

E(X) = p
where E(X) is the expected value of the random variable X, and p is the probability of success.

The variance of the Bernoulli distribution is given by:
Var(X) = p(1 - p)
where Var(X) is the variance of the random variable X, and p is the probability of success.

Binomial Distribution

The binomial distribution is characterized by two parameters: n, the number of trials, and p, the probability of success on a single trial.

The binomial distribution is used to model situations where there are only two possible outcomes on each of n independent trials.

The probability of getting k successes in n trials is given by:

P(X = k) = nCk * pk(1 - p)^(n-k)
where P(X = k) is the probability of getting k successes in n trials, nCk is the binomial coefficient, and p is the probability of success.

The expected value (mean) of the binomial distribution is given by:
E(X) = np
where E(X) is the expected value of the random variable X, n is the number of trials, and p is the probability of success.

The variance of the binomial distribution is given by:
Var(X) = np(1 - p)
where Var(X) is the variance of the random variable X, n is the number of trials, and p is the probability of success.

The Bernoulli distribution is a special case of the binomial distribution where n = 1.

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.

```python
import scipy.stats as stats

# Define the mean and standard deviation
mean = 50
std_dev = 10

# Calculate the z-score for a value of 60
z_score = (60 - mean) / std_dev

# Calculate the probability using the standard normal distribution
probability = stats.norm.cdf(z_score)

print("Probability of a value greater than 60:", probability)
```

Q8: What is the z score? State the importance of the z score.

Answer = A z-score, also known as a standard score, is a measure of how far a particular value is from the mean of a distribution, expressed in units of the standard deviation. It is calculated by subtracting the mean of the distribution from the value in question and then dividing by the standard deviation.

z = (x - μ) / σ
where:

z is the z-score
x is the value in question
μ is the mean of the distribution
σ is the standard deviation of the distribution
Importance of the z-score

The z-score is a valuable statistical tool for several reasons:

Standardizing Data: It allows us to compare values from different data sets by standardizing them to a common scale. This is particularly useful when the data sets have different units or scales of measurement.

Assessing Deviation from Normality: Z-scores help us assess whether a particular value deviates significantly from the norm in a normally distributed data set. A z-score greater than 3 or less than -3 is generally considered an outlier.

Calculating Probabilities: Z-scores can be used to calculate the probability of obtaining a particular value in a normally distributed data set. This is done using the standard normal distribution, which is a standardized version of the normal distribution with a mean of 0 and a standard deviation of 1.

Hypothesis Testing: Z-scores play a crucial role in statistical hypothesis testing. They are used to calculate p-values, which are used to determine whether there is sufficient evidence to reject a null hypothesis.

Predictive Modeling: Z-scores can be used to develop predictive models, such as regression models, to estimate values based on other variables.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Answer = The central limit theorem (CLT) is a fundamental concept in probability theory that describes the behavior of the distribution of sample means as the sample size increases. It states that, for many situations, the sampling distribution of the standardized sample mean tends towards the standard normal distribution even if the original variables themselves are not normally distributed.

Significance of the Central Limit Theorem

The CLT is a significant theorem for several reasons:

Widespread Applicability: The CLT applies to a wide range of random variables, regardless of their underlying distribution. This makes it a versatile tool for analyzing and understanding the behavior of sample means.

Normality Assumption: The CLT allows us to make inferences about populations based on samples, even if the population is not normally distributed. This is a crucial assumption in many statistical tests and procedures.

Error Estimation: The CLT provides a framework for estimating the margin of error in sampling, which is essential for making reliable inferences about populations.

Statistical Inference: The CLT plays a central role in statistical inference, allowing us to construct confidence intervals, perform hypothesis tests, and estimate population parameters.

Applications in Various Fields: The CLT has broad applications in various fields, including finance, engineering, social sciences, and natural sciences. It is used to model phenomena, make predictions, and assess risks in various contexts.

Example of the Central Limit Theorem

Consider the distribution of heights in a large population. The heights of individuals in this population might not be normally distributed, but the distribution of sample means of heights will tend towards the normal distribution as the sample size increases. This means that even if we only have data for a small sample of the population, we can still make inferences about the mean height of the entire population with some confidence.




Q10: State the assumptions of the Central Limit Theorem.

Answer = the Central Limit Theorem:

Independent Observations: The observations in the sample must be independent of each other. This means that the outcome of one observation should not influence the outcome of any other observation.

Identically Distributed Observations: The observations in the sample must be identically distributed. This means that they should all follow the same probability distribution, even if it is not normally distributed.

Large Sample Size: The sample size should be sufficiently large. The exact definition of "large" depends on the specific distribution of the observations, but in general, a sample size of 30 or more is considered large enough for the CLT to apply.

Finite Variance: The observations should have a finite variance. This means that the average squared deviation from the mean should be finite. If the variance is infinite, the CLT will not apply.

Central Limit Property: The distribution of the observations should have a central limit property. This means that the distribution of the sample means will tend towards the standard normal distribution as the sample size increases.
